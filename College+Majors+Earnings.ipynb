# Visualizing Earnings Based on College Majors

**Using visualizations, we can start to explore questions from the dataset like:**

* Do students in more popular majors make more money?
    * Using scatter plots
* How many majors are predominantly male? Predominantly female?
    * Using histograms
* Which category of majors have the most students?
    * Using bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")

recent_grads.iloc[0]
```

```python
#drop missing rows before plotting
raw_data_count = recent_grads.shape[0]

recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = len(recent_grads)

cleaned_data_count
```

# Pandas Scatter Plots

## Sample Size and Median

```python
recent_grads.plot(x = "Sample_size", y = "Median", kind = "scatter", title = "Sample Size vs Median", figsize = (10,5))
```

## Sample Size and Unemployment Rate

```python
recent_grads.plot(x = "Sample_size", y = "Unemployment_rate", title = "Sample Size vs Unemployment Rate", kind = "scatter", figsize = (10, 5))
```

## Full Time and Median

```python
recent_grads.plot(x = "Full_time", y = "Median", kind = "scatter", figsize = (10,5), title = "Full Time vs Median")
```

## Sharewomen and Unemployment Rate

```python
recent_grads.plot(x = "ShareWomen", y = "Unemployment_rate", title = "Full Time vs Unemployment Rate", kind = "scatter", figsize = (10,5))
```

## Man and Median

```python
recent_grads.plot(x = "Men", y = "Median", title = "Man vs Median", kind = "scatter", figsize = (10,5))
```

## Women and Median

```python
recent_grads.plot(x = "Women", y = "Median", title = "Women vs Median", kind = "scatter", figsize = (10,5))
```

# Pandas Histogram

```python
# use for loop to iterate through columns
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize = (10, 12))

for r in range(1, 9):
    ax = fig.add_subplot(4,2,r)
    ax = recent_grads[cols[r-1]].plot(kind = "hist", rot = 40)
```

# Pandas Scatter Plot Matrix

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[["Sample_size", "Median"]], figsize = (10,10))
```

```python
scatter_matrix(recent_grads[["Sample_size", "Median", "Unemployment_rate"]], figsize = (15,15))
```
