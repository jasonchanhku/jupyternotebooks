# Visualizing Gender Gap in College Degrees

## A quick look at some popular college degrees

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

# Comparing Across All Degrees

## Additional improvements
* Removing x axis labels for all except bottom charts
* Added horizontal line at equal gender gap of y = 50

```python
#categorizing the degrees
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize = (16, 16))

#generating first column
for sp in range(0, 18, 3):
    cat_index = int(sp/3)
    ax1 = fig.add_subplot(6, 3, sp + 1)
    ax1.plot(women_degrees["Year"], women_degrees[stem_cats[cat_index]], label = "Women", c = cb_dark_blue, linewidth = 3)
    ax1.plot(women_degrees["Year"], 100 - women_degrees[stem_cats[cat_index]], label = "Men", c = cb_orange, linewidth = 3)

    for i in ["top", "bottom", "left", "right"]:
        ax1.spines[i].set_visible(False)
    
    ax1.tick_params(top = "off", bottom = "off", left = "off", right = "off", labelbottom = "off")
    ax1.set_xlim(1968, 2011)
    ax1.set_ylim(0, 100)
    ax1.set_title(stem_cats[cat_index])
    ax1.set_yticks([0, 100])
    ax1.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)

    if cat_index == 0:
        ax1.text(2003, 85, 'Women')
        ax1.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax1.text(2005, 87, 'Men')
        ax1.text(2003, 7, 'Women')
        ax1.tick_params(labelbottom = "on")
    
#generating second column

for sp in range(1, 16, 3):
    cat_index = int((sp-1)/3)
    ax1 = fig.add_subplot(6, 3, sp + 1)
    ax1.plot(women_degrees["Year"], women_degrees[lib_arts_cats[cat_index]], label = "Women", c = cb_dark_blue, linewidth = 3)
    ax1.plot(women_degrees["Year"], 100 - women_degrees[lib_arts_cats[cat_index]], label = "Men", c = cb_orange, linewidth = 3)

    for i in ["top", "bottom", "left", "right"]:
        ax1.spines[i].set_visible(False)
    
    ax1.tick_params(top = "off", bottom = "off", left = "off", right = "off", labelbottom = "off")
    ax1.set_xlim(1968, 2011)
    ax1.set_ylim(0, 100)
    ax1.set_title(lib_arts_cats[cat_index])
    ax1.set_yticks([0, 100])
    ax1.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)

    if cat_index == 0:
        ax1.text(2003, 85, 'Women')
        ax1.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax1.text(2005, 87, 'Men')
        ax1.text(2003, 7, 'Women')
        ax1.tick_params(labelbottom = "on")

        
#generating third column

for sp in range(2, 18, 3):
    cat_index = int((sp-2)/3)
    ax1 = fig.add_subplot(6, 3, sp + 1)
    ax1.plot(women_degrees["Year"], women_degrees[other_cats[cat_index]], label = "Women", c = cb_dark_blue, linewidth = 3)
    ax1.plot(women_degrees["Year"], 100 - women_degrees[other_cats[cat_index]], label = "Men", c = cb_orange, linewidth = 3)

    for i in ["top", "bottom", "left", "right"]:
        ax1.spines[i].set_visible(False)
    
    ax1.tick_params(top = "off", bottom = "off", left = "off", right = "off", labelbottom = "off")
    ax1.set_xlim(1968, 2011)
    ax1.set_ylim(0, 100)
    ax1.set_title(other_cats[cat_index])
    ax1.set_yticks([0, 100])
    ax1.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)

    if cat_index == 0:
        ax1.text(2003, 85, 'Women')
        ax1.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax1.text(2005, 87, 'Men')
        ax1.text(2003, 7, 'Women')
        ax1.tick_params(labelbottom = "on")

        
plt.savefig("gender_degrees.png")
        
plt.show()





```
